### **Question 1:** What is the difference between multithreading and multiprocessing?

**Answer:**  
Multithreading runs multiple threads within the same process, while multiprocessing runs multiple independent processes with separate memory spaces.

**Differences:**
1. **Memory:** Threads share the same memory; processes have separate memory.  
2. **Performance:** Multithreading suits I/O-bound tasks; multiprocessing suits CPU-bound tasks.  
3. **Overhead:** Threads are lightweight; processes have higher memory and startup overhead.

### **Question 2:** What are the challenges associated with memory management in Python?

**Answer:**  
1. **Garbage collection overhead** – Can slow down performance during cleanup.  
2. **Reference cycles** – Circular references delay memory release.  
3. **Memory fragmentation** – Frequent allocation/deallocation can waste memory.  
4. **Objects not freed** – Long-lived references keep objects in memory.  
5. **Large objects** – Arrays or dataframes may exceed available RAM.

```python
#Question 5: Write a program that handles both IndexError and KeyError using a try-except block.
data = [1, 2, 3]
dictionary = {"name": "Joyesh"}

try:
    print(data[5])           # IndexError
    print(dictionary["age"]) # KeyError
except IndexError:
    print("Index out of range!")
except KeyError:
    print("Key not found in dictionary!")

```

### **Question 6:** What are the differences between NumPy arrays and Python lists?

**Answer:**
1. **Data type:** NumPy arrays are homogeneous; lists can hold mixed types.  
2. **Performance:** NumPy arrays are faster and more memory-efficient.  
3. **Functions:** NumPy supports vectorized and mathematical operations; lists don’t.  
4. **Memory layout:** Arrays use contiguous memory blocks; lists use object references.
### **Question 7:** Explain the difference between apply() and map() in Pandas.
**Explanation:**  
- `map()` works only on Series and applies a function element-wise.  
- `apply()` works on both Series and DataFrames, applying a function along rows or columns.
